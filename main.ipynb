### Barplot

**Dataset Story**

price: price in dollars ($326$18,823)

carat: weight (0.25.01)

cut: quality (Fair, Good, Very Good, Premium, Ideal)

color: color (from J (worst) to D (best))

clarity: cleanliness, clarity (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x: length in mm (010.74)

y: width in mm (058.9)

z: depth in mm (031.8)

depth: percentage of total depth = z / mean(x, y) = 2 * z / (x + y) (4379)

table: width relative to the widest point of the diamond (4395)
